### <b>Task 2:</b> Template Matching using Histograms

```python
import cv2
import matplotlib.pyplot as plt
import pafy
```

#### Template Image

```python
barrel = cv2.imread('barrel.jpg', 1)
plt.imshow(barrel)
```

```python
hsv_barrel = cv2.cvtColor(barrel, cv2.COLOR_BGR2HSV)
plt.imshow(hsv_barrel)
```

#### Obtaining the video using pafy

```python
url = 'https://www.youtube.com/watch?v=A9BVr7kltl8'
vid = pafy.new(url)
play = vid.getbest()
```

```python
capVid = cv2.VideoCapture(play.url)
```

```python
width = int(capVid.get(3))
height = int(capVid.get(4))
size = (width, height)
print(size)
```

```python
fps = capVid.get(cv2.CAP_PROP_FPS)
print(fps)
```

```python
hsvVid = cv2.VideoWriter('vidsection.avi', cv2.VideoWriter_fourcc(*'MJPG'), fps, size)
```

```python
count = 0
```

#### Trimming the video to obtain a clip of few seconds

```python
while capVid.isOpened():
    ret, frame = capVid.read()
    if ret == True:
        count += 1
        if count > 720 and count < 930:
            hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
            hsvVid.write(hsv_frame)
    else:
        break
```

```python
capVid.release()
hsvVid.release()
```

#### Obtaining and normalizing the histograms for the template; all three channels

```python
hist0 = cv2.calcHist([hsv_barrel], [0], None, [256], [0, 256])
hist1 = cv2.calcHist([hsv_barrel], [1], None, [256], [0, 256])
hist2 = cv2.calcHist([hsv_barrel], [2], None, [256], [0, 256])
```

```python
plt.figure(figsize = (10, 3))

plt.subplot(131)
plt.plot(hist0)
plt.title('Hist Channel 0')

plt.subplot(132)
plt.plot(hist1)
plt.title('Hist Channel 1')

plt.subplot(133)
plt.plot(hist2)
plt.title('Hist Channel 2')

plt.show()
```

```python
cv2.normalize(hist0, hist0, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX);
cv2.normalize(hist1, hist1, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX);
cv2.normalize(hist2, hist2, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX);
```

```python
plt.figure(figsize = (10, 3))

plt.subplot(131)
plt.plot(hist0)
plt.title('Norm Hist Channel 0')

plt.subplot(132)
plt.plot(hist1)
plt.title('Norm Hist Channel 1')

plt.subplot(133)
plt.plot(hist2)
plt.title('Norm Hist Channel 2')

plt.show()
```

#### Obtaining the output
###### - loading the trimmed clip
###### - create an object for the output clip
###### - backproject the frames using the histograms from the template image
###### - applying threshold on the three channels
###### - merging the output from thresholding the three channels
###### - bitwise OR between the final and target image

```python
resVid = cv2.VideoCapture('vidsection.avi')
```

```python
width = int(resVid.get(3))
height = int(resVid.get(4))
size = (width, height)
print(size)
```

```python
fps = resVid.get(cv2.CAP_PROP_FPS)
print(fps)
```

```python
outVid = cv2.VideoWriter('output.avi', cv2.VideoWriter_fourcc(*'MJPG'), fps, size)
```

```python
count1 = 0
```

```python
disc = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
```

```python
while resVid.isOpened():
    ret, frame = resVid.read()
    if ret == True:
        backproj0 = cv2.calcBackProject([frame], [0], hist0, [0, 256], scale = 1)
        backproj1 = cv2.calcBackProject([frame], [1], hist1, [0, 256], scale = 1)
        backproj2 = cv2.calcBackProject([frame], [2], hist2, [0, 256], scale = 1)
        
        cv2.filter2D(backproj0, -1, disc, backproj0);
        cv2.filter2D(backproj1, -1, disc, backproj1);
        cv2.filter2D(backproj2, -1, disc, backproj2);
        
        thresh0 = cv2.adaptiveThreshold(backproj0,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,7,4)
        thresh1 = cv2.adaptiveThreshold(backproj1,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,7,4)
        thresh2 = cv2.adaptiveThreshold(backproj2,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,7,4)
        
        fin_thresh = cv2.merge((thresh0, thresh1, thresh2))
        
        result = cv2.bitwise_or(fin_thresh, frame)
        outVid.write(result)
    else:
        break
```

```python
outVid.release()
resVid.release()
```
