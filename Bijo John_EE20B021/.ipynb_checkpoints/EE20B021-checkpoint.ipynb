```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('media/einstein.jpg', 0)
img2 = cv2.GaussianBlur(img, (5, 5), 0)
img3 = cv2.GaussianBlur(img, (9, 9), 0)
diff = img2 - img3
plt.figure()
plt.imshow(diff,cmap='gray'),plt.title("Edge Detection using difference of Gaussians"),plt.axis("off")
plt.show()
```

# Assignment 2

```python
import cv2;

barrel = cv2.imread("media/barrel2.jpg", 1)
barrel = cv2.cvtColor(barrel, cv2.COLOR_BGR2HSV)
histbarrel = cv2.calcHist([barrel],[0, 1], None, [180, 256], [0, 180, 0, 256])
cv2.normalize(histbarrel, histbarrel, 0, 255, cv2.NORM_MINMAX)
vid = cv2.VideoCapture("media/VideoClip.mp4")
while vid.isOpened():
    ret, frame = vid.read()
    hsvframe = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    hsvframe = cv2.calcBackProject([hsvframe],[0,1],histbarrel,[0,180,0,256],1)
    disc = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(5,5))
    hsvframe = cv2.filter2D(hsvframe,-1,disc)
    ret, thresh = cv2.threshold(hsvframe,80,150,cv2.THRESH_BINARY)
    thresh = cv2.merge((thresh,thresh,thresh))
    output = cv2.bitwise_or(frame,thresh)
    cv2.imshow('Template Matching using Histograms',output)
    cv2.waitKey(10)
vid.release()
cv2.destroyAllWindows()
```
