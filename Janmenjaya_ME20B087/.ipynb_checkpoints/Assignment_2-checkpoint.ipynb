## Task 1: Edge Detection using Difference of Gaussians

```python
# go to required directory
%cd ~/cv-session/
```

```python
# Import libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

## ORIGINAL IMAGE

# Reading the image
img = cv.imread('media/einstein.jpg')

# Plotting the image
plt.figure(),plt.axis("off"),plt.title("Original Image"),plt.imshow(img)
plt.show()

## GAUSSIAN BLUR AND THEIR DIFFERENCE IN ORIGINAL IMG

# Gaussian blur with different kernels
gauss1 = cv.GaussianBlur(img,(5,5),sigmaX=1,sigmaY=1)
gauss2 = cv.GaussianBlur(img,(9,9),sigmaX=10,sigmaY=10)

# Difference of two gaussian fiters
diff = gauss1-gauss2

# Plotting the image
plt.figure(),plt.axis("off"),plt.title("Edge_Detection (Without Denoising)"),plt.imshow(diff)
plt.show()

## GAUSSIAN BLUR AND THEIR DIFFERENCE IN DENOISED IMG

# Noise reduction of img
b,g,r = cv.split(img)
rgb_img = cv.merge([r,g,b])

dst = cv.fastNlMeansDenoisingColored(img,None,15,15,7,21)
b,g,r=cv.split(dst)
dust=cv.merge([r,g,b])

# Gaussian blurs with different kernels
blur1 = cv.GaussianBlur(dust,(5,5),sigmaX=1,sigmaY=1)
blur2 = cv.GaussianBlur(dust,(9,9),sigmaX=10,sigmaY=10)

# Noise reduction in gaussian blur
b,g,r = cv.split(blur1)
rgb_img1 = cv.merge([r,g,b])

dst1 = cv.fastNlMeansDenoisingColored(blur1,None,20,20,7,21)
b,g,r=cv.split(dst1)
dust1=cv.merge([r,g,b])

b,g,r = cv.split(blur2)
rgb_img2 = cv.merge([r,g,b])

dst2 = cv.fastNlMeansDenoisingColored(blur2,None,20,20,7,21)
b,g,r=cv.split(dst2)
dust2=cv.merge([r,g,b])

# Difference of two Gaussain filters
edge = dust1-dust2

# Plotting the image
plt.figure(),plt.axis("off"),plt.title("Edge_Detection (After Denoising)"),plt.imshow(edge)
plt.show()
```

## Task 2: Template Matching using Histograms

```python
# Go to the required directory
%cd ~/Downloads

# Import the library
import cv2 as cv
import matplotlib.pyplot as plt

# Reading the image
barrel_img = cv.imread('barrel.png')

# The conversion
hsv_barrel = cv.cvtColor(barrel_img, cv.COLOR_BGR2HSV)

# Calculation of histogram
barrel_hist = cv.calcHist([hsv_barrel], [0, 2], None, [20, 256], [0, 20 ,0 ,256])

# histogram normalisation
cv.normalize(barrel_hist, barrel_hist, 0, 255, cv.NORM_MINMAX)

# Video capture
capture = cv.VideoCapture('video.mp4')

ref = True

while(capture.isOpened()):
    
    # Obtain the present frame
    ref, frame = capture.read()
    
    # Conversion of frame to hsv
    hsv_frame = cv.cvtColor(frame, cv.COLOR_BGR2HSV)
    
    # Calculation of backprojection
    back_proj = cv.calcBackProject([hsv_frame], [0, 1], barrel_hist, [0, 20, 0, 256], 1)

    # Creation of convoluted disk element
    convol = cv.getStructuringElement(cv.MORPH_ELLIPSE, (8, 8))
    
    # Filteration of back projection using convolution
    cv.filter2D(back_proj, -1, convol, back_proj) 

    # Application of threshold
    _1, thresh = cv.threshold(back_proj, 80, 160, cv.THRESH_BINARY)

    # Merging of thersholds
    merged = cv.merge((thresh, thresh, thresh))

    final = cv.bitwise_and(frame, merged)
    final = cv.erode(final, (9, 9))
    final = cv.erode(final, (9, 9))
    final = cv.dilate(final, (7, 7))
    final = cv.GaussianBlur(final, (9, 9), 9)
    final = cv.cvtColor(final, cv.COLOR_BGR2GRAY)

    # Achieve the contours
    conts = cv.findContours(final.copy(), cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)[-2] 

    # Management of rectange size
    for k in conts:
        if cv.contourArea(k) > 300:
            (a, b, c, d) = cv.boundingRect(k)
            cv.rectangle(frame, (a, b), (a + c, b + d), (255, 0, 0), 2)

    # show the frames
    cv.imshow("IGVC", frame)

    if cv.waitKey(30) == 27:
        break

cv.destroyAllWindows()
capture.release()
```
