```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread(r'C:\Users\kapil\Desktop\einstein.jpg')
```

```python
blur_gaus_1 = cv2.GaussianBlur(img,(5,5),0)
blur_gaus_2 = cv2.GaussianBlur(img,(9,9),2)
Edge_Detected = blur_gaus_1 - blur_gaus_2
```

```python
plt.figure(),plt.axis("off"),plt.title("Original Image"),plt.imshow(img)
plt.figure(),plt.axis("off"),plt.title("Edge Detected Image"),plt.imshow(Edge_Detected)
plt.show()
```
