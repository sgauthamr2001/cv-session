```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
!wget https://github.com/sgauthamr2001/cv-session/raw/main/media/einstein.jpg
img = cv2.imread('einstein.jpg')
```

```python

cv2_imshow(img)
```

```python
blur_gauss1 = cv2.GaussianBlur(img, (5, 5), 0)
blur_gauss2 = cv2.GaussianBlur(img, (9, 9), 0)

plt.figure(),plt.axis("off"),plt.title("Original Image"),plt.imshow(img)
plt.figure(),plt.axis("off"),plt.title("Blurred Image 1"),plt.imshow(blur_gauss1)
plt.figure(),plt.axis("off"),plt.title("Blurred Image 2"),plt.imshow(blur_gauss2)
plt.show()
```

```python
# Edge detection by subtracting the 2 gaussian filters
cv2_imshow(blur_gauss1 - blur_gauss2)
```
