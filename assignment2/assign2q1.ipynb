```python
#question1.
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread('einstein.jpg')
blur1 = cv.GaussianBlur(img,(5,5),0)
blur2 = cv.GaussianBlur(img,(9,9),0)
plt.figure(),plt.axis('off'),plt.title("Original Image"),plt.imshow(img)
plt.figure(),plt.axis("off"),plt.title("5X5 blured"),plt.imshow(blur1)
plt.figure(),plt.axis("off"),plt.title("9x9 blured"),plt.imshow(blur2)
res95 = blur2 - blur1
plt.figure(),plt.axis("off"),plt.title("9-5 final"),plt.imshow(res95)
res59 = blur1 - blur2
plt.figure(),plt.axis("off"),plt.title("5-9 final"),plt.imshow(res59)
plt.show()
```
