```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

img = cv2.imread('einstein.jpg',1)

plt.imshow(img)
```

```python
blur1=cv2.GaussianBlur(img,(5,5),0)
blur2=cv2.GaussianBlur(img,(9,9),0)
plt.imshow(blur1)
```

```python
plt.imshow(blur2)
```

```python
edge = blur2 - blur1
plt.imshow(edge)
```
