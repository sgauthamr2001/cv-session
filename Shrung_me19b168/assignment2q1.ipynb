```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('../media/einstein.jpg',1)
plt.figure(figsize=(8,8)),plt.axis("off"),plt.title("Einstein"),plt.imshow(img);
```

```python
gaus_blur_5 = cv2.GaussianBlur(img, (5, 5), 0)
gaus_blur_9 = cv2.GaussianBlur(img, (9, 9), 0)

# Showing the blurred images
fig, ax = plt.subplots(1,2, figsize=(24,12))
ax[0].axis("off")
ax[0].set_title('Gaussian Blurring with 5x5 kernel')
ax[0].imshow(gaus_blur_5)

ax[1].axis("off")
ax[1].set_title('Gaussian Blurring with 9x9 kernel')
ax[1].imshow(gaus_blur_9);
```

```python
new_img = cv2.subtract(gaus_blur_9, gaus_blur_5)
new_img_2 = gaus_blur_9 - gaus_blur_5
new_img_3 = gaus_blur_5 - gaus_blur_9

fig, ax = plt.subplots(1,3, figsize=(24,12))
ax[0].axis("off")
ax[0].set_title('Edge detection using cv2.subtract()')
ax[0].imshow(new_img)

ax[1].axis("off")
ax[1].set_title('Edge detection by subtracting image arrays')
ax[1].imshow(new_img_2)

ax[2].axis("off")
ax[2].set_title('Edge detection by subtracting image arrays')
ax[2].imshow(new_img_3)
```

```python
hist = cv2.calcHist([img] , [0] , None , [256] , [0,256])
plt.figure(),plt.plot(hist),plt.xlabel('Pixel intensity value'),plt.ylabel('Frequency'),plt.title('Histogram');

hist_blur1 = cv2.calcHist([gaus_blur_5] , [0] , None , [256] , [0,256])
plt.figure(),plt.plot(hist_blur1),plt.xlabel('Pixel intensity value'),plt.ylabel('Frequency'),plt.title('hist_blur1');

hist2 = cv2.calcHist([new_img_2] , [0] , None , [256] , [0,256])
plt.figure(),plt.plot(hist2),plt.xlabel('Pixel intensity value'),plt.ylabel('Frequency'),plt.title('Hist2');
```
